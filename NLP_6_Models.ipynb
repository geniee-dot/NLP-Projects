```python
# ============================================
# NLP CLASSIFICATION - ALL 6 MODELS
# ============================================

# Install & imports
import pandas as pd # work with tables/dataframes (like Excel in Python)
import numpy as np # mathematical operations on arrays/numbers
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer # converts text → word counts # converts text → weighted word importance
from sklearn.model_selection import train_test_split # splits data into training and testing sets
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

print("All libraries loaded!")
```

STEP 2: LOAD DATA

```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd

# We'll use 2 categories only (binary sentiment-like classification)
# Corrected 'rec.sport.football' to 'rec.sport.baseball' as it exists in the dataset
categories = ['rec.sport.baseball', 'sci.med']

# Load training and test data
train_data= fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
test_data= fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))

# converting to dataframe
df_train = pd.DataFrame({'text': train_data.data, 'label': train_data.target})
df_test= pd.DataFrame({'text': test_data.data, 'label': test_data.target})

# checking the data
# Corrected df_train(10) to df_train.head(10) to display the first 10 rows of the DataFrame
print(df_train.tail(10))
print("training sample :", len(df_train))
print("testing sample :" , len(df_test))

for i, name in enumerate(train_data.target_names):
  print(f" {i} ={name}")
```

```python
from sklearn.datasets import fetch_20newsgroups

# load everything first
all_data = fetch_20newsgroups(subset='train')

# print all category names
print(all_data.target_names)
```

# STEP 3 - VECTORIZATION

```python
# separarte text and labels
x_train = df_train['text']
y_train = df_train['label']

x_test = df_test['text']
y_test = df_test['label']

# convert text to numbers using TF-IDF
vectorizer =TfidfVectorizer(stop_words='english')# automatically removes english stop wored eg(the,is,am,are...)
x_train_vec = vectorizer.fit_transform(x_train)
x_test_vec = vectorizer.transform(x_test)

print("The training shape:", x_train_vec.shape) # shape-gives u the size of of the table(rows and column) of the trained vector
print("The testing shape:", x_test_vec.shape)


```

X is capital   =   usually a table (many columns)
y is lowercase =   usually one column (labels)

# STEP 4 - TRAIN ALL 6 MODELS

PS- BECAUSE WE ARE TRAINING ( WE USE THE FIT FUNCTION)

```python
# Model 1 - Naive Bayes:

nb = MultinomialNB()
nb.fit(x_train_vec, y_train)

# Model 2- Logistic Regression
lr = LogisticRegression()
lr.fit(x_train_vec, y_train)

# Model 3- SVM : Support vector machines
svm = LinearSVC()
svm.fit(x_train_vec, y_train)

# Model 4- Random Forest
rf = RandomForestClassifier()
rf.fit(x_train_vec, y_train)

# Model 5 - Decison Tree
dt= DecisionTreeClassifier()
dt.fit(x_train_vec, y_train)

# Model 6 - KNN
knn = KNeighborsClassifier()
knn.fit(x_train_vec, y_train)


```

STEP 5 - PREDICTIONS & ACCURACY

```python
nb_pred = nb.predict(x_test_vec)
lr_pred = lr.predict(x_test_vec)
svm_pred = svm.predict(x_test_vec)
rf_pred = rf.predict(x_test_vec)
dt_pred = dt.predict(x_test_vec)
knn_pred = knn.predict(x_test_vec)

# Check accuracy of all 6 models
print("Model Accuracies:")
print("Naive Bayes:         ", metrics.accuracy_score(y_test, nb_pred))
print("Logistic Regression: ", metrics.accuracy_score(y_test, lr_pred))
print("SVM:                 ", metrics.accuracy_score(y_test, svm_pred))
print("Random Forest:       ", metrics.accuracy_score(y_test, rf_pred))
print("Decision Tree:       ", metrics.accuracy_score(y_test, dt_pred))
print("KNN:                 ", metrics.accuracy_score(y_test, knn_pred))
```
